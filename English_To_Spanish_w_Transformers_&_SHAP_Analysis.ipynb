# Text to Text Explanation: Machine Translation w/ Transformers

This notebook demonstrates use of generating model explanations for a text to text scenario on a pretrained transformer model for machine translation. In this demo, we showcase explanations on 2 different models provided by Hugging Face which inclues translation from English to Spanish (https://huggingface.co/Helsinki-NLP/opus-mt-en-es).

```python
import numpy as np
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import shap
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-es")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-es")
```

```python
data = ["Transformers are quickly become the model of choice over NLP problems, because Attention is all you need!"]
```

```python
explainer = shap.Explainer(model, tokenizer)

shap_values = explainer(data, fixed_context=1)
```

```python
# View the i/o heatmap
shap.plots.text(shap_values)
```
